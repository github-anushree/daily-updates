# Data Transformation

```python
#31. Remove the EmployeeID column from the DataFrame.
df = df.drop('EmployeeID', axis = 1)
```

```python
#verify
df.head()
```

```python
#32. Reset the index of the DataFrame.
df.reset_index()
```

```python
#33. Replace all employees who did not achieve the target with "Needs Improvement" in the TargetAchieved column.
df['TargetAchieved'] = df['TargetAchieved'].replace('No', 'Needs Improvement')
```

```python
#verify
df[df['TargetAchieved'] == 'Needs Improvement']
```

```python
#verify
df['TargetAchieved'].value_counts()
```

```python
# 34. Create a new column PerformanceLevel with values: "High" if SalesAmount > 40000, "Medium" if between 20000-40000, else "Low".
df['PerformanceLevel'] = np.where (
    df['SalesAmount'] > 40000, 'High', 
    np.where (df['SalesAmount'] >= 20000, 'Medium', 'Low')
)
```

```python
df.head()
```

```python
#verify
df['PerformanceLevel'].value_counts()
```

```python
#35 Convert Age column to an integer type.
df['Age'] = df['Age'].astype(int)
```

```python
#verify
df['Age'].dtype
```

```python
#36. Create a new column AboveAverageEarnings, which is "Yes" if TotalEarnings is above the average.
df['AboveAverageEarnings'] = df['TotalEarnings'] > df['TotalEarnings'].mean()

#This will create a column with True/False values.
```

```python
#verify
df[['AboveAverageEarnings', 'TotalEarnings']].head()
```

```python
#or check out
df['AboveAverageEarnings'].value_counts()
```

```python
#37. Convert the Department column to a categorical data type.
df['Department'] = df['Department'].astype('category')
```

```python
#verify
df['Department'].dtype
```

```python
# 38. Find the top 5 employees based on total earnings.
df.sort_values('TotalEarnings', ascending = False).head(5)
```

```python
#39. Find the bottom 5 employees based on total earnings.
df.sort_values('TotalEarnings', ascending = True).head(5)
```

```python
##If you want to select only certain columns:
df[['Name', 'Age', 'Department']].sort_values('TotalEarnings', ascending = True).head(5)
```

```python
#OR--
df.sort_values('TotalEarnings').head(5)  #ascending = True will be deafult, nahi lagay to bhi chalega
```
