```python
#Multiple except blocks
try:
    x = 10 / 0
except ZeroDivisionError:
    print('Can not divided by zero')
except NameError:
    print('The variable does not exist')
```

| Term                | Meaning                                |
| ------------------- | -------------------------------------- |
| `ZeroDivisionError` | Error type → dividing by zero          |
| `NameError`         | Error type → variable not defined      |
| Specific except     | Catch only that type of error          |
| Simple except       | Catches everything but not recommended |


🟩 Final Answer (super simple)

ZeroDivisionError and NameError are specific Python error types.
They are not mandatory, but they are used to catch specific mistakes.
This helps you write better, safer, cleaner code.

```python
#Using else and finally (Beginner-friendly)
#✔ else → runs only when no error happens
try:
    x = 5 / 1
except ZeroDivisionError:
    print('Error Can not divided by zero')
else:
    print('Everything went fine')
```

```python
#✔ finally → runs no matter what
#Useful for: closing files, closing database connections, etc.
try:
    x = 5 / 0
except ZeroDivisionError:
    print('Divided by zero')
finally:
    print('This always runs')
```

```python
try:
    x = 5 / 0
except:
    print("Error happened")
finally:
    print("This always runs")
```

1️⃣4️⃣ Handle errors in list to numeric conversion

```python
#Loop through list items and convert each to int. Use try/except to ignore bad values.
my_list = ['5', '4', '9', '1', 'hi', 'hello', 'chi', '10']
converted_list = []


for item in my_list:
    try:
        num = int(item)  # try converting
        converted_list.append(num)
    except ValueError:
        print(f'Skipping bad value: {item}')

print('Converted list:', converted_list)
        
```

1️⃣5️⃣ Catch errors during merging two DataFrames

```python
#If merge column doesn't exist, catch the exception.
import pandas as pd

try:
    df1 = pd.DataFrame({'id' : [1, 2, 3], 'name': ['A', 'B', 'C']})
    df2 = pd.DataFrame({'ID' : [1, 2, 3], 'salary':[100, 200, 300]}) # column name mismatch
    # Try merging on 'id'
    result = pd.merge(df1, df2, on = 'id')  # KeyError will occur here

except KeyError:
    print('Error: the merge column does not exists in one of the dataframes')
else:
    print('Merge successful!')
    print(result)
```

1️⃣6️⃣ Use try/except inside a loop

```python
#Loop through a list of values. Convert each to int. Count how many conversions failed.
my_list = ['1', '2', '3', '4', 'hi', 'x']
converted_list = []
fail_count = 0

for item in my_list:
    try:
        num = int(item)  # try converting
        converted_list.append(num)  # success

    except ValueError:
        fail_count = fail_count + 1 # count failures

print('Converted list', converted_list)
print('Failed conversion', fail_count)    
```
