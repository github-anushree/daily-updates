# Data Types (List, Tuple, Set, Dict)

```python
#1- Given a list of numbers, remove duplicates without changing order.
l = [4, 5, 1, 2, 0, 0, 4, 4]
#convert it into set
s = set(l)
print(s)
```

```python
#without changing order
l = [4, 5, 1, 2, 0, 0, 4, 4]

lst = []
for o in l:
    if o not in lst:
        lst.append(o)
print(lst)
```

```python
#2- Find the first non-repeating element in a list.
l = [4, 5, 1, 2, 0, 4]

lst = []
for item in l:
    if l.count(item) == 1:
        lst.append(item)
        break
print(lst)
        
```

```python
# 3- Given two lists, return common elements using:
#List
#Set (optimized approach)
```

```python
#using list
l1 = [4, 5, 1, 2, 0, 4]
l2 = [2, 4, 6, 8]

common = []

for item in l1:
    if item in l2 and item not in common:
        common.append(item)

print(common)
```

```python
#using set
l1 = [4, 5, 1, 2, 0, 4]
l2 = [2, 4, 6, 8]

common = list(set(l1) & set(l2))

common
```

```python
#4- Convert this list into a dictionary:
keys = ["name", "age", "city"]
values = ["Anushree", 24, "Delhi"]

d = dict(zip(keys, values))
d
```

```python
#loop method
keys = ["name", "age", "city"]
values = ["Anushree", 24, "Delhi"]

item = {}
for i in range(len(keys)):
    item[keys[i]] = values[i]
print(item)
```

```python
# 5- Given dictionary:
#data = {"A": 100, "B": 200, "C": 300}
#Find key with maximum value.

data = {"A": 100, "B": 200, "C": 300}

result = max(data, key= data.get)
print(result)
```

```python
#using lambda
data = {"A": 100, "B": 200, "C": 300}

result = max(data, key = lambda k: data[k])
result
```

```python
#6- Find second highest number in list:
lst = [10, 20, 4, 45, 99]

#if duplicate exist
unique_numbers = set(lst)
unique_numbers.remove(max(set(lst)))
sec_highest = max(unique_numbers)
sec_highest
```

```python
#7- Count frequency of characters in string:
c = "dataanalyst"

freq = {}

for char in c:
    if char in freq:
        freq[char] += 1
    else:
        freq[char] = 1

print(freq)
```

```python
# 8- Return dictionary with number as key and frequency as value.
lst = [1,2,2,3,3,3,4]

freq = {}

for num in lst:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1
print(freq)
```

# List Comprehension

```python
# 1- Create list of squares from 1 to 20.

sq_lst = [x**2 for x in range(1, 21)]

print(sq_lst)
```

```python
#normal loop
sq_list = []

for i in range(1, 21):
    sq_list.append(i**2)
print(sq_list)
```

```python
# 2- From this list:
nums = [10, 15, 20, 25, 30]
# Create new list containing only numbers divisible by 10.
lst = [x for x in nums if x % 10 == 0]
print(lst)
```

```python
# find output
x = 5

for i in range(1, 6):
    if i % 2 == 0:
        continue
    x = x + i
    x = x * 2
print(x)
```

```python
# 3- Flatten this nested list using list comprehension:
lst = [[1,2], [3,4], [5,6]]

l = [item for sublist in lst for item in sublist]
print(l)
```

```python
#using normal loop
lst = [[1,2], [3,4], [5,6]]
flat_list = []

for sublist in lst:
    for item in sublist:
        flat_list.append(item)
print(flat_list)
```

```python
#4 - Create dictionary using comprehension:
#d = {1:1, 2:4, 3:9, 4:16}

com = {x : x**2 for x in range(1, 5)}
com
```

```python
#Reverse key-value:
d = {1:1, 2:4, 3:9, 4:16}

rev = {v:k for k, v in d.items()}
rev
```

```python
lst1 = [1, 2]
lst2 = [1, 2]
for x in lst1:
    lst2.append(x)
for y in lst2:
    print(y)
```

```python
# 5- Convert list of strings to uppercase using comprehension.
lst = ['abc', 'efg', 'xyz']

upper = [x.upper() for x in lst]
upper
```
