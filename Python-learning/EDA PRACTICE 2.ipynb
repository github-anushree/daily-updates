## 8️⃣ SECTION 7 — Merge / Join

```python
#Create another dataframe.

customers = pd.DataFrame({
    'customer' :  ['A', 'B', 'C'],
    'segment' : ['Premium', 'Standard', 'Premium']
})

df2 = customers
print(df2)
```

```python
#51- Concatenate two dataframes vertically.
pd.concat([df, df2])
#⚠️ Technically correct but logically mismatched
```

```python
#Reason:
#df and df2 have different columns, so vertical concat will produce many NaN values.
pd.concat([df, df2], axis = 0)
```

```python
#52- Concatenate horizontally.
pd.concat([df, df2], axis = 1)
```

```python
#53- Merge df and customers using customer
df.merge(df2, on = 'customer')
#This performs an inner join by default.
```

```python
#OR--
pd.merge(df, df2, on= 'customer')
```

```python
#54- Perform inner join
pd.merge(df, df2, how = 'inner', on = 'customer')
#Inner join returns only matching rows.
```

```python
#55- Perform outer join
pd.merge(df, df2, how = 'outer', on = 'customer')
#Outer join returns all rows from both tables.
```

```python
#56- Perform left join
pd.merge(df, df2, how = 'left', on = 'customer')
#Left join keeps all rows from left dataframe (df).
```

```python
#57- Perform right join
pd.merge(df, df2, how = 'right', on = 'customer')
#Right join keeps all rows from right dataframe (df2).
```

```python
#Find sales and customer segment together.
df.merge(customers, on = 'customer', how = 'left')
```

```python
#3️⃣ Join using different column names
pd.merge(df, df2, left_on = 'customer', right_on = 'customer')
```
