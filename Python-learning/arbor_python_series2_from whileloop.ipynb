```python
#6- Write a function to compute the nth Fibonacci number.
#Description: Implement a function that returns the nth Fibonacci number, where Fibonacci sequence starts with 0 and 1, and each 
#subsequent number is the sum of the previous two.
def fibonacci(n):
    a = 0
    b = 1

    if n == 0:
        return a
    if n == 1:
        return b

    for i in range(2, n+1):
        c = a + b
        a = b
        b = c
    return b

print(fibonacci(7))
    
```

```python
#9- Write a function to calculate the factorial of each number in a given list.
#Description: Create a function that accepts a list of numbers as input and returns a new list containing the factorial of each number 
#in the original list.
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result


def factorial_list(numbers):
    result_list = []
    for num in numbers:
        result_list.append(factorial(num))
    return result_list

print(factorial_list([2, 3, 4, 5, 6]))
```

```python
#Question number 5: Create a function to generate a list of prime numbers within a specified range.
#Description: Define a Python function that takes two parameters, start and end, representing the range, and returns a list of prime 
#numbers within that range.
def prime(start, end):
    prime_list = []

    for num in range(start, end + 1):
        if num > 1:
            is_prime = True

            for i in range(2, int(num**0.5) + 1):
                if num % i == 0:
                    is_prime = False
                    break

            if is_prime:
                prime_list.append(num)

    return prime_list

print(prime(2, 10))
```

```python
#Question number 6: Design a function to calculate the exponentiation of a number to a given power.
#Description: Implement a Python function that accepts two parameters, base and exponent, and returns the result of raising the base 
#to the power of the exponent.
def power(base, exponent):
    result = 1

    for i in range(exponent):
        result = result * base
    return result
print(power(2, 5))
    
```

```python
#Question number 7: Write a function to remove duplicate elements from a list.
#Description: Develop a Python function that takes a list as input and returns a new list with duplicate elements removed, 
#preserving the original order.
def remove_duplicate(a):
    unique_list = []

    for item in a:
        if item not in unique_list:
            unique_list.append(item)

    return unique_list

print(remove_duplicate([4, 4, 5, 5, 6, 6, 10, 2, 3, 7, 8, 9, 9]))
```

```python
#Question number 10: Write a function to find the mode (most frequent element) in a list.
#Description: Develop a Python function that takes a list of elements as input and returns the mode (the element that appears most frequently)
#in the list. If there are multiple modes, return any one of them.
def find_mode(lst):
    mode = lst[0]

    for item in lst:
        if lst.count(item) >  lst.count(mode):
            mode = item
    return mode
print(find_mode([2, 2, 2, 2, 3, 3, 4, 5, 6, 6, 6, 6, 8 ,9, 6, 9, 10, 6, 4]))
            
```

```python
# Question 12:
# Create a class named Student with attributes name, roll_number, and grades.
# Include methods to add grades, calculate the average grade, and display the student's information.
# Create an instance of the Student class with name "John", roll number "12345", and grades [85, 90, 88].
# Add another grade (95), calculate the average grade, and display the student's information.
class Student:
    def __init__(self, name, roll_number, grades):
        self.name = name
        self.roll_number = roll_number
        self.grades = grades

    def add_grade(self, grade):
        self.grades.append(grade)

    def calculate_average(self):
        return sum(self.grades) / len(self.grades)

    def display_info(self):
        print("Name:", self.name)
        print("Roll Number:", self.roll_number)
        print("Grades:", self.grades)
        print("Average Grade:", self.calculate_average())


# Create object
p = Student("John", "12345", [85, 90, 88])

# Add new grade 95
p.add_grade(95)

# Display all info
p.display_info()
```

```python
# Question 13:
# Define a class called Triangle with attributes side1, side2, and side3.
# Include methods to check if the triangle is valid, calculate its perimeter, area, and display its sides.
# Create an instance of the Triangle class with side lengths 3, 4, and 5, then check if it is valid, calculate its perimeter and area,
#and display its sides.
import math  # needed for square root

class Triangle:
    def __init__ (self, side1, side2, side3):
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3

    def is_valid(self):
        # Triangle inequality: sum of any two sides > third side
        return (self.side1 + self.side2 > self.side3 and
               self.side1 + self.side3 > self.side2 and
               self.side2 + self.side3 > self.side1)

    def perimeter(self):
        return self.side1 + self.side2 + self.side3

    def area(self):
        # Using Heron's formula
        s = self.perimeter() / 2
        return math.sqrt(s * (s - self.side1) * (s - self.side2) * (s - self.side3))

    def display_sides(self):
        print('Sides of triangle: ', self.side1, self.side2, self.side3)


# Create object
t = Triangle(3, 4, 5)

# Check if valid
if t.is_valid():
    print("Triangle is valid")
    print("Perimeter:", t.perimeter())
    print("Area:", t.area())
    t.display_sides()
else:
    print("Triangle is not valid")
```

‚úÖ What does ‚Äúcategorical data type‚Äù mean in pandas?

A categorical data type in pandas is used for columns that have a limited number of fixed, repeating categories.

Example categories:

"Petrol"

"Diesel"

"Electric"

These categories repeat many times.

‚úÖ Why convert to categorical type?
1. Saves memory

Pandas stores each category only once and uses integer codes internally.

2. Speeds up operations

Grouping, sorting, filtering becomes faster.

3. Makes data cleaner

It tells pandas:
‚ÄúThis column contains categories, not free-text.‚Äù

| Type            | Meaning                              |
| --------------- | ------------------------------------ |
| **object**      | plain text                           |
| **int / float** | numeric values                       |
| **category**    | text with repeated, fixed categories |

iloc = ‚ÄúGive me the first row.‚Äù

loc ‚Üí label
iloc ‚Üí index
